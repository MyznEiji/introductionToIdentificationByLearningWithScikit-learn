#  #3 Hold-out2

```python
import numpy as np

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

X = data.data
y = data.target

from sklearn import linear_model
clf = linear_model.LogisticRegression()

from sklearn.model_selection import ShuffleSplit

ss = ShuffleSplit(n_splits=1, train_size=0.5, test_size=0.5)

# trainとtestを分割 : hold-out
train_index, test_index = next(ss.split(X, y))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
ss = ShuffleSplit(n_splits=10, train_size=0.5, test_size=0.5)

for train_index, test_index in ss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
```

```python
ss = ShuffleSplit(n_splits=10, train_size=0.95, test_size=0.05)

train_index, test_index = next(ss.split(X, y))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]
```

```python
# トレーニングデータとテストデータで比率が保たれない
print(np.unique(y, return_counts=True))
print(np.unique(y, return_counts=True)[1] / y.size)
print(np.unique(y_train, return_counts=True)[1] / y_train.size)
print(np.unique(y_test, return_counts=True)[1] / y_test.size)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
ss = StratifiedShuffleSplit(n_splits=1, train_size=0.95, test_size = 0.05, random_state=0)
```

```python
train_index, test_index = next(ss.split(X, y))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]
```

```python
# トレーニングデータとテストデータで比率が保たれている
print(np.unique(y, return_counts=True))
print(np.unique(y, return_counts=True)[1] / y.size)
print(np.unique(y_train, return_counts=True)[1] / y_train.size)
print(np.unique(y_test, return_counts=True)[1] / y_test.size)
```

# #4 cross validation

```python
import numpy as np

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

X = data.data
y = data.target

from sklearn import linear_model
clf = linear_model.LogisticRegression()

# cross validation
from sklearn.model_selection import KFold
ss = KFold(n_splits=10, shuffle=True)

for train_index, test_index in ss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
    
    
```

```python
print(np.unique(y, return_counts=True))
print(np.unique(y, return_counts=True)[1] / y.size)
```

```python
# データの比率が変わってしまっている
ss = KFold(n_splits=10, shuffle=True)

for train_index, test_index in ss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    print(np.unique(y_train, return_counts=True)[1] / y_train.size, y_train.size,
          np.unique(y_test, return_counts=True)[1] / y_test.size, y_test.size)
```

```python
from sklearn.model_selection import StratifiedKFold
ss = StratifiedKFold(n_splits=10, shuffle=True)

for train_index, test_index in ss.split(X, y):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train,  y_test = y[train_index], y[test_index]
    

    print(np.unique(y_train, return_counts=True)[1] / y_train.size, y_train.size,
          np.unique(y_test, return_counts=True)[1] / y_test.size, y_test.size)
```

```python
# cross_validationをしてscoreを計算するオブジェクト

from sklearn.model_selection import cross_val_score
# cv = StratifiedKFold
ave_score = cross_val_score(clf, X, y, cv = 10)
```

```python
print("{0:4.2f} +/-  {1:4.2f} %".format(ave_score.mean() * 100, ave_score.std() * 100) )
```

# #5 Leave one out

```python
import numpy as np

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

X = data.data
y = data.target

from sklearn import linear_model
clf = linear_model.LogisticRegression()
```

```python
# leave-one-outモジュールをインポート　
# テストデータを一つだけ取り出して、１つでデータの数だけ検証する
from sklearn.model_selection import LeaveOneOut
loocv = LeaveOneOut()
```

```python
train_index, test_index = next(loocv.split(X, y))
```

```python
y.size, train_index.size, test_index.size
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, y, cv=loocv)
```

```python
scores.mean() * 100, scores.std() * 100, scores.size
```

```python
from sklearn.model_selection import LeavePOut
# テストデータを2つ抜き出して他を学習
loocv = LeavePOut(2)
```

```python
# leave one-group out
group = np.array(list(range(50)) * 12)
group = np.sort(group[:y.size])
group.size
```

```python
group
```

```python
from sklearn.model_selection import LeaveOneGroupOut
loocv = LeaveOneGroupOut()

for train_index, test_index in loocv.split(X, y, group):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
    
```

```python
# cross validationでもできる

from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, y, groups=group, cv=loocv)
```

```python
scores.mean(), scores.std(), scores.size
```

# #5 学習データ・検証データ・テストデータ

```python
import numpy as np

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

X = data.data
y = data.target

from sklearn import linear_model
clf = linear_model.LogisticRegression()

from sklearn.model_selection import ShuffleSplit
ss = ShuffleSplit(n_splits=1, train_size=0.8, test_size=0.2, random_state=0)

train_index, test_index = next(ss.split(X, y))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X_train, y_train, cv=10)
```

```python
scores.mean()
```

```python
# Cのパラメータの値の準備
C_range_exp = np.linspace(start=-15, stop=20, num=36)
C_range = 10 ** C_range_exp
print(C_range)
```

```python
# どのパラメータがいいのかを検証
all_scores_mean = []
all_scores_std  =[]

for C in C_range:
    clf.C = C
    scores = cross_val_score(clf, X_train, y_train, cv=10)
    
    all_scores_mean.append(scores.mean())
    all_scores_std.append(scores.std())
```

```python
all_scores_mean = np.array(all_scores_mean)
all_scores_std = np.array(all_scores_std)
```

```python
# Cのパラメーターはどのあたいが　いいのかを検証　
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(C_range_exp, all_scores_mean)

plt.errorbar(C_range_exp, all_scores_mean, yerr=all_scores_std)
```

```python
all_scores_mean.max()
```

```python
max_index = np.argmax(all_scores_mean)
```

```python
C_range_exp[max_index]
```

```python
clf.C = 10 ** C_range_exp[max_index]
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
# デフォルトの値でも検証
clf = linear_model.LogisticRegression()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
